```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.cross_validation import StratifiedKFold
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# function to create model
def create_nn():
    # create model
    model = Sequential()
    model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
    model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
    
    # compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
# load pima indians dataset
df = pd.read_csv("diabetes.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop('Outcome', axis=1).as_matrix()
Y = df.Outcome.as_matrix()
```

```python
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn=create_nn, epochs=100, batch_size=10)))
```

```python
pipeline = Pipeline(estimators)
```

```python
# evaluate using 10-fold cross validation
kfold = StratifiedKFold(y=Y, n_folds=5, shuffle=True, random_state=seed)
```

```python
results = cross_val_score(pipeline, X, Y, cv=kfold)
print(results.mean())
```

```python
print(results.mean())
```

```python
rfc = RandomForestClassifier()
```

```python
forest_results = cross_val_score(rfc, X, Y, cv=kfold)
```

```python
print(forest_results.mean())
```

The neural network model performs better than the standard random forest across 10 folds in the data.

Random Forests tend to work very well right out of the box. The data does not need to be scaled. Random Forests work quite well on low dimensional data. EEnsemble works to reduce overfitting and decrease result variance.

Neural Networks are much more complex than random forests, requiring more intensive computation. However, they learned better.
