```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('results.csv')
pd.set_option('display.max_columns', None)
```

```python
df = raw_data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
for item in ['25k', '10k', 'half', '30k', '5k', '20k', '35k', '40k']:
    df[item] = df[item].apply(lambda s: s.replace('-', ''))
```

```python
for item in ['25k', '10k', 'half', '30k', '5k', '20k', '35k', '40k']:
    df[item] = pd.to_numeric(df[item])
```

```python
for item in ['25k', '10k', 'half', '30k', '5k', '20k', '35k', '40k', 'official', 'pace']:
    df[item] = pd.to_timedelta(df[item], unit='m')
```

```python
g = sns.countplot('gender', data=df)
g.set_title("Participants by Gender")
```

```python
male = df[df['gender'] == 'M']
female = df[df['gender'] == 'F']
```

```python
print(male['official'].mean())
print(female['official'].mean())
```

```python
print(male['overall'].mean())
print(female['overall'].mean())
```

```python
df['male'] = np.where(df['gender'] == 'M', 1, 0)
```

```python
g = sns.countplot('age', data=df, palette="cubehelix")
g.figure.set_size_inches(13,8)
g.set_title("Participants per age group")
```

```python
g = sns.countplot('age', data=df, palette="cubehelix", hue='gender')
g.figure.set_size_inches(13,6)
g.set_title("Participants per Age & Gender")
```

```python
g = sns.boxplot(df['gender'], df['age'], palette="cubehelix")
g.figure.set_size_inches(13,6)
g.set_title("Distribution of finish times per Age group")
```

```python
g = sns.lmplot(x='overall', y='age', data=df, hue='gender', palette="cubehelix", size=10)
```

```python
s = df.groupby('country').count()['city'].sort_values(ascending=False).head(20)
g = sns.barplot(s.index, s, palette='rainbow')
g.figure.set_size_inches(13,8)
g.set_title("Most popular Country")
```

```python
s = df.groupby('country').count()['city'].sort_values(ascending=False).head(21)[1:]
g = sns.barplot(s.index, s, palette='rainbow')
g.figure.set_size_inches(13,8)
g.set_title("Most popular Country (after US)")
```

```python
s = df.groupby('country').count()['city'].sort_values(ascending=False).head(22)[2:]
g = sns.barplot(s.index, s, palette='rainbow')
g.figure.set_size_inches(13,8)
g.set_title("Most popular Country (after US and Canada)")
```

```python
s = df[df['country'] == 'USA'].groupby('state').count()['country'].sort_values(ascending=False).head(20)
g = sns.barplot(s.index, s, palette="rainbow")
g.figure.set_size_inches(13,8)
g.set_title("Most popular State")
```

```python
g = sns.boxplot(df['age'], df['official'].dt.seconds, palette="coolwarm")
g.figure.set_size_inches(13,8)
g.set_title("Distribution of finish times per Age group")
```

```python
g = sns.boxplot(male['age'], male['official'].dt.seconds, palette="coolwarm")
g.figure.set_size_inches(13,8)
g.set_title("Distribution of male finish times per Age group")
```

```python
g = sns.boxplot(female['age'], female['official'].dt.seconds, palette="coolwarm")
g.figure.set_size_inches(13,8)
g.set_title("Distribution of female finish times per Age group")
```

```python
df['official_secs'] = df['official'].dt.seconds
g = sns.pointplot(x='age', y='official_secs', hue='gender', data=df)
g.figure.set_size_inches(13,8)
g.set_title("Official Time Splits by Age and Gender")
```

```python
df['half_secs'] = df['half'].dt.seconds
g = sns.pointplot(x='age', y='half_secs', hue='gender', data=df)
g.figure.set_size_inches(13,8)
g.set_title("Halfway Time Splits by Age and Gender")
```

```python
df['pace_secs'] = df['pace'].dt.seconds
g = sns.pointplot(x='age', y='pace_secs', hue='gender', data=df)
g.figure.set_size_inches(13,8)
g.set_title("Pace Splits by Age and Gender")
```
